# Data Visualization - Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Simple Line Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
y2 = np.cos(x)
plt.plot(x, y, label='sin')
plt.plot(x, y2, label='cos')
plt.legend()
plt.show()
```
